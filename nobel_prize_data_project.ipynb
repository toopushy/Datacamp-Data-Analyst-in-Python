The Nobel Prize has been among the most prestigious international awards since 1901. Each year, awards are bestowed in chemistry, literature, physics, physiology or medicine, economics, and peace. In addition to the honor, prestige, and substantial prize money, the recipient also gets a gold medal with an image of Alfred Nobel (1833 - 1896), who established the prize.

![](Nobel_Prize.png)

The Nobel Foundation has made a dataset available of all prize winners from the outset of the awards from 1901 to 2023. The dataset used in this project is from the Nobel Prize API and is available in the `nobel.csv` file in the `data` folder.

In this project, you'll get a chance to explore and answer several questions related to this prizewinning data. And we encourage you then to explore further questions that you're interested in!

```python
# Loading in required libraries
import pandas as pd
import seaborn as sns
import numpy as np

# Corrected file path
df = pd.read_csv("data/nobel.csv")

top_gender = df['sex'].value_counts().index[0]
print(top_gender)
top_country = df['birth_country'].value_counts().index[0]
print(top_country)

df['is_usa'] = df['birth_country'] == 'United States of America'
df['decade'] = (np.floor(df['year']/10)*10).astype(int)

df.head()

max_decade_usa = df.groupby('decade', as_index = False)['is_usa'].mean()
max_decade_usa = max_decade_usa[max_decade_usa['is_usa'] == max_decade_usa['is_usa'].max()]['decade'].values[0]

print(max_decade_usa)

df['is_female'] = df['sex'] == 'Female'

max_female = df.groupby(['decade', 'category'], as_index = False)['is_female'].mean()
#max_female.head()

max_female = max_female[max_female['is_female'] == max_female['is_female'].max()][['decade', 'category']].values[0]

#print(max_female)

max_female_dict = {max_female[0]: max_female[1]}
print(max_female_dict)

new_df = df[df['is_female']]
#new_df.head()

first_woman_name_list = new_df[new_df['year'] == new_df['year'].min()][['full_name']].values[0]
delimiter = " "
first_woman_name = delimiter.join(first_woman_name_list)

first_woman_category_list = new_df[new_df['year'] == new_df['year'].min()][['category']].values[0]
delimiter = " "
first_woman_category = delimiter.join(first_woman_category_list)


print(first_woman_name)
print(first_woman_category)

counts = df['full_name'].value_counts()
counts = counts[counts >= 2].index
repeat_list = list(counts)
print(repeat_list)
```
